###  ¿Cuántos Usuarios gustan del Jazz?

```redshift
SET search_path TO tickit;

SELECT COUNT(*) 
FROM users 
WHERE likejazz = true;
```

###  ¿Cuántos Usuarios gustan de la ópera y del rock al mismo tiempo?

```redshift
SELECT COUNT(*) 
FROM users 
WHERE likeopera = true AND likerock = true;
```

### ¿Cuál es el promedio, moda y mediana del total de Ventas?

```redshift
-- Promedio del total de ventas
SELECT AVG(s.pricepaid) AS promedio_ventas
FROM sales s;

-- Mediana del total de ventas
SELECT
  pricepaid AS mediana_ventas
FROM (
  SELECT
    pricepaid,
    ROW_NUMBER() OVER (ORDER BY pricepaid) AS row_num,
    COUNT(*) OVER () AS total_rows
  FROM sales
) subquery
WHERE row_num = (total_rows + 1) / 2
   OR row_num = total_rows / 2 + 1;

-- Moda del total de ventas
SELECT
  pricepaid AS moda_ventas
FROM (
  SELECT
    pricepaid,
    COUNT(*) AS frequency
  FROM sales
  GROUP BY pricepaid
  ORDER BY frequency DESC
  LIMIT 1
) subquery;
```

### ¿Cuál el promedio de ventas de usuarios que gustan del rock, pero NO del Jazz?

```redshift
SELECT AVG(s.pricepaid) AS promedio_ventas_rock
FROM sales s
JOIN users u ON s.buyerid = u.userid
WHERE u.likerock = true AND u.likejazz = false;
```
